```python
"""
Given semantic recon loss, find the token count needed to achieve that. 
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
resnet_fintuned_nll = [1.3031, 1.7386, 2.2132, 2.3732, 2.4852, 2.5706, 2.4659, 2.3595, 1.9322]
mlp_recon_loss_only = [1.6380, 1.8619, 2.0249, 2.1416, 2.1866, 2.1995, 2.1802, 2.1246, 1.9898]
mlp_with_dino_dist = [1.3616, 1.6165, 1.8659, 2.0430, 2.1415, 2.1845, 2.2169, 2.1127, 1.9626]
```

```python
plt.figure(figsize=(10, 4))
k_values = [1, 2, 4, 8, 16, 32, 64, 128, 256]
plt.plot(k_values, resnet_fintuned_nll, marker='o', label='ResNet Finetuned')
plt.plot(k_values, mlp_recon_loss_only, marker='o', label='MLP Recon Loss Only')
plt.plot(k_values, mlp_with_dino_dist, marker='o', label='MLP with DINO Dist')
plt.xscale('log', base=2)
plt.xlabel('Token Count (k)')
plt.ylabel('Semantic Recon Loss (NLL)')
plt.title('Semantic Recon Loss vs Token Count')
plt.legend()
plt.grid(True, which="both", ls="--")
plt.savefig("/BS/data_mani_compress/work/thesis/thesis/output/predicting_token_count/dino.pdf")
plt.show()
```

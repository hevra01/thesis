```python
import json
import numpy as np
import matplotlib.pyplot as plt
import torch
import os
```

```python
# read density json file
# main file: /BS/data_mani_compress/work/thesis/thesis/data/datasets/density_imagenet/val/reconst_
# loop over all reconst_*.json files
k = [1,2,4,8,16,32,64,128,256]
print("starting")
for i in k:
    combined_json = []
    # for file in this directory
    # loop over all files in the directory
    print("hi")
    dir_path = f"/BS/data_mani_compress/work/thesis/thesis/data/datasets/density_imagenet/val/reconst_{i}/token_count256"
    for file in sorted(os.listdir(dir_path)):
        print(f"Processing file: {file}")
        with open(os.path.join(dir_path, file), 'r') as f:
           data = json.load(f)
        combined_json.extend(data)
        print(len(combined_json))
    if len(combined_json) != 50000:
        print(f"Warning: combined json length is {len(combined_json)} instead of 50000")
        break
    # save combined json
    with open(f"/BS/data_mani_compress/work/thesis/thesis/data/datasets/density_imagenet/val/reconst_{i}_combined_token_count256.json", 'w') as f:
        json.dump(combined_json, f)
```

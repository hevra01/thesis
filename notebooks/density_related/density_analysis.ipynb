```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "/BS/data_mani_compress/work/thesis/thesis/data/datasets/density_imagenet/original_val/unconditional/combined.json"
with open(file,'r') as f:
    current_densities = json.load(f)
    
```

```python
print(len(current_densities))
print(len(current_densities[0]), current_densities[0])
```

```python
# Stack the 50-element inner lists from the first element of each entry: shape (50000, 50)
inner = np.array([item[0] for item in current_densities])  # (N, 50)

# Sum across all images for each of the 50 positions
position_sums = inner.sum(axis=0)  # (50,)

plt.figure(figsize=(10, 4))
plt.bar(np.arange(len(position_sums)), position_sums)
plt.xlabel('Position (index)')
plt.ylabel('Sum across all images')
plt.title('Sum of density values at each of the 50 positions')
plt.tight_layout()
plt.show()
```

```python
# Average the second element (a scalar) across all 50k images
second_element_avg = np.mean([item[1] for item in current_densities])
print(f"Average of second element across all images: {second_element_avg:.6f}")
```
